```python
import pandas as pd

ts_ms = pd.date_range(start='2025-01-01',
                    end='2025-10-02',
                    periods=100,
                    tz='Asia/Seoul')
data = {'date': ts_ms.strftime('%Y-%m-%d')}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df.date)
# df.info()
# 년도를 추출해서 새로운 컬럼으로 넣기
df['year'] = df.date.dt.year
df['month'] = df.date.dt.month
df['day_int'] = df.date.dt.weekday

def convert_week(x):
    temp = ['월','화','수','목','금','토','일']
    return temp[x]

df['day_str'] = df['day_int'].apply(lambda x : convert_week(x))

# 주말(토일)은 1 나머지는 0
def convert_weekend(x):
    if x == '토' or x == '일':
        return 1
    return 0
df['day_weekend'] = df['day_str'].apply(lambda x : 1 if x in ['토','일'] else 0)
df.head()
```
