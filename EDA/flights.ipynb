```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.pivot(index='year', columns='month', values='passengers')
```

```python
import matplotlib.pyplot as plt
# 연도별 승객수의 변화 lineplot - sns
# sns.lineplot(data=flights, x='year', y='passengers', errorbar=('ci', 95))
# sns.lineplot(data=flights, x='year', y='passengers', errorbar=None)
sns.lineplot(data=flights, x='year', y='passengers')
# sns 시각화는 신뢰구간도 같이 표시하는 경우
# 데이터가 변동될수 있는 구간
# 신뢰구간이 넓으면 불확실성이 크다 (데이터 분산이 크거나 샘플수가 적음)
# 신뢰구간이 좁으면 안정적 (데이터가 모여있거나 샘플수가 많을 경우)
# plt.legend(loc='best')
plt.show()
```
